## 02 - Tratar as colunas sem informação adicionando o texto “Não informado”.

### Importando bibliotecas

```python
import numpy as np
import pandas as pd
```

```python
# Lendo o arquivo csv 
df = pd.read_csv('topyoutube.csv')

df
```

```python
#Filtrando Artist sem informações
linhas_faltando_informacao = df[df.isnull().any(axis=1)]
print(linhas_faltando_informacao)
```

```python
#Criando um nova dataframe calterando dados nulos para "Não Informados"
df_sem_Nulos =  df.fillna("Não Informado")
df_sem_Nulos
# Salvando novo CSV sem dados nulos
df_sem_Nulos.to_csv("topyoutube_SemNulos.csv")
df_sem_Nulos
```

## 03 - Formatar a coluna de total de inscritos multiplicando por 100 milhões.

```python
# Trocando o Nome da Coluna de 100M para "Total de Inscritos" 
df_sem_Nulos.rename(columns={'100M': 'Total de Inscritos'}, inplace=True)
df_sem_Nulos
```

```python
df_sem_Nulos['Total de Inscritos'] = df_sem_Nulos['Total de Inscritos'].apply(lambda x: x * 100 if x != 'Não Informado' else x)
```

```python
df_sem_Nulos
```

```python
# Salvando para o arquivo 
df_sem_Nulos.to_csv("topyoutube_X100.csv")
```

## 04 - Formatar coluna “AVG” para 2 casas decimais depois da vírgula.

```python
# Formatando o a Coluna Avg para 2 Casas
df_sem_Nulos['Avg'] = df['Avg'].apply(lambda x: '{:.2f}'.format(x))
```

```python
df_sem_Nulos
```

```python
# Salvando para o arquivo 
df_sem_Nulos.to_csv("topyoutube_AvgFormatado.csv")
```

## 05 - Mostrar os top 10 usuários.

```python
# Removendo linhas com "Não Informado" na coluna 'Total de Inscritos'
df_sem_nao = df_sem_Nulos[df_sem_Nulos['Total de Inscritos'] != 'Não Informado']
```

```python
df_ordenado_Mais_Incritos = df_sem_nao.sort_values(by='Total de Inscritos', ascending=False)
```

```python
df_ordenado_Mais_Incritos.head(10)
```

## 06 - Mostrar primeiros 100 usuários por nome decrescente.

```python
df_ordenado_Nome = df_sem_nao.sort_values(by='Artist', ascending=True)
```

```python
df_ordenado_Nome.head(100)
```
