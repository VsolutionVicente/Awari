## **Objetivo**

Nesta prática, vamos aprender sobre outra biblioteca para geração de gráficos, que complementa e enriquece recursos de Matplotlib, chamada **Seaborn**.

image.png

- É uma biblioteca Python usada para plotar gráficos com a ajuda de Matplotlib, Pandas e Numpy.
- É construído sobre o Matplotlib e é considerado um superconjunto (*superset*) da biblioteca Matplotlib.
- Ajuda na visualização de dados univariados e bivariados.
- Usa belos temas para decorar gráficos Matplotlib.
- Atua como uma ferramenta importante na representação de Modelos de Regressão Linear.
- Serve para fazer gráficos de dados estáticos de séries temporais, elimina a sobreposição de gráficos e também auxilia no seu embelezamento.

Documentação: https://seaborn.pydata.org/

## **Prática**

Vamos repetir os gráficos que fizemos em Matplotlib, mas agora com o Seaborn, apenas para vermos as diferenças e semelhanças.

```python
# Importando bibliotecas
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns #além do matplotlib, importamos também o Seaborn.
```

### **Gráfico de linha**

#### **Matplotlib**

```python
np.random.seed(42)
x = np.arange(10)
y = np.random.normal(size=10)

plt.plot(x, y)
plt.title("Exemplo")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.show()
```

#### **Seaborn**

Usando *datasets* de exemplo embutidos no próprio Seaborn:

```python
flights = sns.load_dataset("flights")
sns.lineplot(data=flights, x="year", y="passengers");
```

### **Grafico de barras**

#### **Matplotlib**

```python
np.random.seed(42)
x = np.arange(14)
y = np.arange(14)

plt.bar(x, y)
plt.title("Exemplo")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.show()
```

#### **Seaborn**

```python
tips = sns.load_dataset("tips")
sns.barplot(data=tips);
```

### **Histogramas**

#### **Matplotlib**

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins = [0,20,40,60,80,100])
plt.title("histogram")
plt.show()
```

#### **Seaborn**

```python
penguins = sns.load_dataset("penguins")
sns.displot(penguins, x="flipper_length_mm");
```

### **Gráfico de setores**

#### **Matplotlib**

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.title("Exemplo")

plt.show()
```

#### **Seaborn**

Seaborn não tem uma função padrão para criar gráficos de pizza, mas podemos usar a seguinte sintaxe no Matplotlib para criar um gráfico de pizza e adicionar uma paleta de cores Seaborn:

```python
tips = sns.load_dataset("tips")
colors = sns.color_palette('pastel')[0:5]
plt.pie(tips['sex'].value_counts(), labels=tips['sex'].value_counts().keys(), colors = colors, autopct='%.0f%%');
```

### **Boxplot**

#### **Matplotlib**

```python
np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize =(10, 7))

plt.boxplot(data)
plt.title('Exemplo')

plt.show()
```

#### **Seaborn**

```python
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips);
```

### **Scatter plot**

#### **Matplotlib**

```python
# gerar valores demonstrativos
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.title('Exemplo')
plt.show()
```

#### **Seaborn**

```python
# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip");
```
